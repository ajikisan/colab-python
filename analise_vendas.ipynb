Trabalhando com Planilhas do Excel

```python
#Importando a biblioteca com Planilhas do Excel
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("Vendas2010.xlsx")
df2 = pd.read_excel("Vendas2011.xlsx")
df3 = pd.read_excel("Vendas2012.xlsx")

```

```python
#Juntando todos os arquivos excel
df = pd.concat([df1,df2,df3])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 últimas linhas
df.tail()
```

```python
#Verificando linhas aleatórias
df.sample(5)
```

```python
#Verificando o tipo de dados de cada coluna
df.dtypes
```

```python
#Alterando o tipo de dado da coluna
df["Quantidade"] = df["Quantidade"].astype("float64")
```

```python
#Verificando a alteração do tipo de dado
df.dtypes
```

```python
df.head()
```

```python
#Verificando a média do valor de desconto
df["Valor Desconto"].mean()
```

Tratando valores Faltantes

```python
#Consultado linhas com valores faltantes
df.isnull().sum()
```

```python
#Caso houvesse valores nulos, pode-se realizar a substituição na memória pela média
df["Valor Desconto"].fillna(df["Valor Desconto"].mean(),inplace=True)
```

```python
#Amostragem
df.sample(15)
```

```python
#Após execução do código - verificar se ainda existe valores nulos.
df.isnull().sum()
```

```python
#Substituição dos valores nulos por zero
df["Quantidade"].fillna(0,inplace=True)
```

```python
#Apagar as linhas com valores nulos
df.dropna(inplace=True)
df.head(10)
```

```python
#Apagar as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Valor Desconto"], inplace=True)
df.sample(3)
```

```python
#Remover linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
df.sample(5)
```

Criando Novas Colunas

```python
#Criando a Coluna Receita
df["Receita"] = df["Quantidade"].mul(df["Preço Unitário"])
df.head()
```

```python
#Quantidade de Produtos Vendidos
df["Receita/Preço Unitário"] = df["Receita"] / df["Preço Unitário"]
df.sample(5)
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest - Traz 3 vendas com base na coluna receita
df.nlargest(3,"Receita")
```

```python
#nsmallest - Traz 3 vendas com menor receita
df.nsmallest(3,"Receita")
```

```python
#Agrupamento de Receita por Item
df.groupby("Item")["Receita"].sum()
```

```python
#Ordenando o Conjunto de Dados
df.sort_values("Receita", ascending=False).head(10)
```

Trabalhando com Datas

```python
#Transformação da coluna de Data em tipo inteiro
df["Data Venda"] = df["Data Venda"].astype("int64")
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data venda em tipo datetime
df["Data Venda"] = pd.to_datetime(df["Data Venda"])
```

```python
#Verificando o tipo de dados
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data Venda"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano da Venda"] = df["Data Venda"].dt.year
```

```python
#Demosntração aleatória
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["Mês da Venda"], df["Dia da Venda"] = (df["Data Venda"].dt.month, df["Data Venda"].dt.day)
```

```python
#Verificando por amostragem
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data Venda"].min()
```

```python
#Calculando a diferença de dias
df["diferenca de dias"] = df["Data Venda"] -df["Data Venda"].min()
```

```python
#Amostra aleatória
df.sample(5)
```

```python
#Criando a coluna de trimestre
df["Trimestre de Venda"] = df["Data Venda"].dt.quarter
```

```python
#Amostragem
df.sample(5)
```

```python
#Filtrando as vendas do mês de março em 2011
vendas_marco_2011 = df.loc[(df["Data Venda"].dt.year == 2011) & (df["Data Venda"].dt.month == 3)] 
```

```python
# Verificando a variável criada
vendas_marco_2011
```

```python
#Amostragem com a variável criada
vendas_marco_2011.sample(7)
```

Visualização de Dados

```python
#Value Counts Realiza a contagem dos preços unitários do maior para o menor valor
df["Preço Unitário"].value_counts(ascending=False)
```
